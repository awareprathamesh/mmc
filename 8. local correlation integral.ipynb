```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform

%matplotlib inline
```

```python
# Generate random 2D data
X = np.random.rand(100, 2)
```

```python
# Distance matrix
dist_matrix = squareform(pdist(X))
```

```python
# Radius values
r_values = np.linspace(0.01, 1, 50)

C_r = []

for r in r_values:
    count = np.sum(dist_matrix < r) - len(X)  # remove self-count
    C_r.append(count / (len(X)*(len(X)-1)))
```

```python
plt.plot(r_values, C_r)

plt.xlabel("Radius (r)")
plt.ylabel("C(r)")
plt.title("Local Correlation Integral")

plt.show()
```
