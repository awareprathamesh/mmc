```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

%matplotlib inline
```

```python
# Normal data
X1 = np.random.randn(100, 2)

# Add outliers
X2 = np.random.uniform(low=-6, high=6, size=(15, 2))

# Combine
X = np.vstack((X1, X2))
```

```python
dist_matrix = cdist(X, X)
```

```python
# Mean distance from all other points
influence_score = np.mean(dist_matrix, axis=1)
```

```python
# Threshold (top 10% as outliers)
threshold = np.percentile(influence_score, 90)

outliers = X[influence_score > threshold]
normal = X[influence_score <= threshold]
```

```python
plt.scatter(normal[:,0], normal[:,1], color='blue', label='Normal')
plt.scatter(outliers[:,0], outliers[:,1], color='red', label='Outliers')

plt.title("Influenced Outlierness")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()

plt.show()
```
